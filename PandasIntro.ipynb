## Pandas Intro - Fast

First we start off by importing *pandas* and *numpy*. We can import pandas and numpy as anything we want, but *pd* and *np* are chosen as convention

`%matplotlib inline` is used to have plots show up within Jupyter notebook instead of an external viewer

```python
import pandas as pd
import numpy as np

%matplotlib inline
```

## Create a pandas DataFrame
Multiple ways:  
1) Usually import a .csv with `pd.read_csv`  
2) Or create a new instance from scratch, usually from a dictionary, as shown in the code block below

```python
x = np.linspace(-np.pi, np.pi, 201)
y = np.sin(x)
data = {'angle(rad)' : x, 'sin(x)' : y}
df = pd.DataFrame(data)
```

### Plotting and Visualization
Multiple ways to plot in Python  
1) `matplotlib` is the default library for python. import with `import matplotlib.pyplot as plt`  
2) Plot with pandas. Uses `matplotlib` as a back-end. Trades customizability for ease  
3) `seaborn` uses matplotlib. import with `import seaborn as sns`. Trades customizability for aesthetics and ease

```python
import matplotlib.pyplot as plt
plt.plot(df['angle(rad)'], df['sin(x)'])
plt.xlabel('Angle [rad]')
plt.ylabel('sin(x)')
```

```python
df.boxplot()
```

```python
import seaborn as sns
sns.boxplot(df) # won't work due to the way data is formatted right now
```

```python
df_long = pd.melt(df) # "melt" the data from wide format to long format
df_long
```

```python
sns.boxplot(x="variable", y="value", data=df_long)
```
